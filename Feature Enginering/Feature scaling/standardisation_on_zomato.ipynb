```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('enhanced_zomato_dataset_clean.csv')
df.sample(5)
```

```python
df_new = df[[ 'Avg_Rating_Cuisine', 'Avg_Price_Cuisine','Is_Highly_Rated']]
```

```python
df_new.shape
df_new.sample(5)
df['Is_Highly_Rated'].dtype
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(df_new.drop('Is_Highly_Rated',axis=1),df_new['Is_Highly_Rated'], test_size=0.3, random_state=0)
x_train.shape
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python

scaler = StandardScaler()
scaler.fit(x_train)
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
scaler.mean_
```

```python
x_train_scaled=pd.DataFrame(x_train_scaled ,columns=x_train.columns)
x_test_scaled =pd.DataFrame(x_test_scaled ,columns=x_test.columns)
```

```python
np.round(x_train_scaled.describe(),1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr_scaled=LogisticRegression()
lr.fit(x_train,y_train)
lr_scaled.fit(x_train_scaled,y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred_scaled = lr_scaled.predict(x_test_scaled)

from sklearn.metrics import accuracy_score

print(accuracy_score(y_test, y_pred))
print(accuracy_score(y_test, y_pred_scaled))
```

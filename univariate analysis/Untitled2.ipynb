```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('realistic_linear_regression_dataset.csv')
```

```python
df.sample(10)
```

```python
def marketing_analysis(region,category):
    mask=df['Region']==region
    mask2=df['ProductCategory']==category
    dfN=df[mask &mask2]
    sns.boxplot(dfN['MarketingSpend'])
    # sns.distplot(dfN['MarketingSpend'])
```

```python
marketing_analysis('East','Furniture')
```
